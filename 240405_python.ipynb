5.4 반복문에서 사용하는 continue / break

- continue

- break : 반복문 끝내기

```python
# 1~100 홀수, continue

for i in range(1,101):
  if i%2 == 0:
    continue
  print(i,end=',')

# 들여쓰기 내어쓰기 주의
# continue는 if문에 걸림
```

```python
# 1~100 짝수, continue

for i in range(1,101):
  if i%2 == 1:
    continue
  print(i,end=',')

# 들여쓰기 내어쓰기 주의
# continue는 if문에 걸림
```

```python
# 1~100 홀수, break

for i in range(5):
  if i%2 == 1:
    break
  print(i)

# 들여쓰기 내어쓰기 주의
# break는 if문에 걸림
# 홀수가 나오는 순간 브레이크가 걸리는 것
```

```python
# 1에서 10까지, 짝수끼리의 합, 홀수끼리의 합

짝수합 = 0
홀수합 = 0

# 짝수합과 홀수합을 정의 후

for i in range(1,11):
  if i%2 == 0:
    짝수합+= i
  else:
    홀수합+= i

print(짝수합)
print(홀수합)
```

1부터 n까지의 합

```python
n = int(input())

tot=0
for i in range(1,n+1):
  tot+=i

print(tot)


# n의 값을 입력
# 1~n까지 반복
   #합을 계산
# 합을 출력
```
