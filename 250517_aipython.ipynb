### **1. 선 그래프 : plot()**

- plot(x 값, y값, 선스타일 문자열, marker='데이터포인트마커기호', label='그래프이름')

```python
x = [1,2,3,4]  # x축 데이터
y = [1,4,9,16]  # y축 데이터
plt.plot(x,y)
plt.show()
```

3. 막대 그래프 : bar()/barh()

- bar(x값, y값, color='색 문자' 혹은 '색깔 문자열'): 범주별 데이터를 나타내는 세로 막대 그래프 생성

- barh(x값, y값, color='색 문자' 혹은 '색깔 문자열'): 범주별 데이터를 나타내는 가로 막대 그래프 생성

- color 값은 생략 가능

```python
x = ['A', 'B', 'C']
y = [100, 150, 120]

plt.bar(x,y)
plt.show()
```

```python
data_city = pd.DataFrame({'도시':['서울', '부산', '인천', '대구'],
                          '인구 수' : [9904312, 3448737, 2890451, 2466052]})

df3 = data_city
df3
```

```python
plt.rc('font', family='NanumGothic')  # 폰트 설치!!!
plt.bar(df3['도시'], df3['인구 수'])
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')

plt.show()
```

## **3. subplots(행 개수, 열 개수, figsize=(가로 길이, 세로 길이)**



```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

## **4. 빈도그래프(히스토그램) : hist()**

## **2. 막대그래프 : countplot()**

- countplot(x='x축에 범주로 표시할 칼럼명', hue='데이터 구분 칼럼명', data=데이터프레임)

## **5. pairplot()**

- pairplot(데이터프레임, hue='범주를 갖는 칼럼명')

## **6. 산점도와 추체선(회쉬선) 그래프 : regplot()**

- regplot(x='x축에 표시할 칼럼명', y='y축에 표시할 칼럼명', 데이터프레임)

## **7. 집계한 값에 따라 색깔을 표시하는 그래프: heatmap()**

- heatmap(data=2차원 형태로 전달된 데이터, annot=True/False, fmt='파이썬 서식 문자열', cbar=True/False

```python
df=iris_data.corr()
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
```
