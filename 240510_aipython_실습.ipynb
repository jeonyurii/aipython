```python
import pandas as pd

# 데이터를 이용하여 데이터프레임 생성

df = pd.DataFrame({'학생':['John', 'Alice', 'Bob', 'Emily'],    ## 딕셔너리 형태
                   '수학':[90, 87, 78, 92],
                   '영어': [85, 91, 80, 88],
                   '과학': [88, 89, 85, 82]})
```

```python
df
```

```python
# 생성한 데이터의 요약 통계 정보 출력

df.describe()
```

```python
# 생성한 데이터프레임에서 '수학'점수가 가장 높은 학생의 정보 출력

df[df['수학']==df['수학'].max()]
```

```python
# 생성한 데이터프레임에서 '수학'점수가 90점 이상인 학생의 정보 출력

df[df['수학']>=90]
```
